### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA OPCIONAL: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños.

```python
def scatter_plot(df, col_x, col_y, col_cat=None, size=50, escala=1):
    """
    Pinta un diagrama de dispersión flexible.

    Parámetros
    
    df       : pd.DataFrame
    col_x    : str  – columna para el eje X
    col_y    : str  – columna para el eje Y
    col_cat  : str o None  – columna categórica para colorear (opcional)
    size     : int/float o str
                 · número  --> tamaño fijo para todos los puntos
                 · str     --> nombre de columna; sus valores controlan el tamaño
    escala   : float – multiplicador del tamaño cuando size es una columna
    """
    #Calcular tamaños
    if isinstance(size, str):
        sizes = df[size].values * escala
    else:
        sizes = size

    fig, ax = plt.subplots(figsize=(9, 6))

    if col_cat is not None:
        categorias = df[col_cat].unique()
        palette = sns.color_palette('tab10', len(categorias))
        for cat, color in zip(categorias, palette):
            mask = df[col_cat] == cat
            sz = sizes[mask] if isinstance(size, str) else sizes
            ax.scatter(
                df.loc[mask, col_x],
                df.loc[mask, col_y],
                s=sz, color=color, alpha=0.7, label=str(cat), edgecolors='none'
            )
        ax.legend(title=col_cat, bbox_to_anchor=(1.01, 1), loc='upper left')
    else:
        ax.scatter(
            df[col_x], df[col_y],
            s=sizes, alpha=0.7, edgecolors='none'
        )

    ax.set_xlabel(col_x)
    ax.set_ylabel(col_y)
    ax.set_title(f'{col_y} vs {col_x}')
    plt.tight_layout()
    plt.show()

```

### #1.3

Pinta el diagrama de dispersión de longitud y latitud (longitud en las x, latitud en las y) de las ciudades, usando el tamaño de las ciudades para el tamaño de los puntos y los colores para mostrar la población. Previamente crea una columna nueva "log_poblacion" y asignale el log decimal de la poblacion (np.log10() y como argumento los valores de la columna "population_total"). Usa "log_poblacion" para el color.

Una vez calculado el logaritmo, también puedes probar a categorizar la población, aquí tienes un ejemplo de como podría salir para que lo tengas de referencia:

```python
#Crear columna log_poblacion
df['log_poblacion'] = np.log10(df['population_total'].values)
print('Ejemplo de log_poblacion:')
display(df[['city', 'population_total', 'log_poblacion']].head())
```

```python
#Scatter: sin columna categórica,
# tamaño = área total, color interno controlado por log_poblacion mediante colormap

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(
    df['longd'],
    df['latd'],
    s=df['area_total_sq_mi'] * 0.5,
    c=df['log_poblacion'],
    cmap='YlOrRd',
    alpha=0.7,
    edgecolors='grey',
    linewidths=0.3
)

cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('log₁₀(Población)', fontsize=11)

ax.set_xlabel('Longitud', fontsize=12)
ax.set_ylabel('Latitud', fontsize=12)
ax.set_title('Ciudades de California: posición geográfica, área y población', fontsize=13)

#Leyenda tamaños
for area, lbl in [(10, '10 sq mi'), (100, '100 sq mi'), (500, '500 sq mi')]:
    ax.scatter([], [], s=area * 0.5, c='gray', alpha=0.6, label=lbl)
ax.legend(title='Área total', loc='lower right', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
#categorizar la población en cuartiles
df['cat_poblacion'] = pd.qcut(
    df['log_poblacion'], q=4,
    labels=['Muy pequeña', 'Pequeña', 'Media', 'Grande']
)

scatter_plot(
    df,
    col_x='longd',
    col_y='latd',
    col_cat='cat_poblacion',
    size='area_total_sq_mi',
    escala=0.5
)
```

En esta parte tendrás que hacer dos cosas, crear una presentación sobre el titanic de no más de 6 slides, sin incluir portada, con los siguientes mensajes y su apoyo en datos:

3- ... que la clase en la que se viajes tuvo una influencia significativa y podría explicar...  
4- ... las diferencias de tasas de supervivencia entre mujeres y niños.    
5- Además aunque el puerto de embarque parece influir, se podría asumir que lo fue el tipo de pasaje que embarcó en cada uno.  

Y generar el notebook con las gráficas que vas a emplear en la presentación. Tienes que entregar ambas cosas.

```python
titanic = pd.read_csv('./data/titanic.csv')
titanic.dropna(subset=['survived', 'pclass', 'sex', 'age', 'embarked'], inplace=True)
print(f'Registros válidos: {len(titanic)}')
display(titanic.head())
```

```python
#Diapositiva 1: Tasa de mortalidad global
fig, ax = plt.subplots(figsize=(5, 4))
surv_rate = titanic['survived'].mean()
cats = ['Fallecidos', 'Supervivientes']
vals = [1 - surv_rate, surv_rate]
colors = ['#E63946', '#2A9D8F']
bars = ax.bar(cats, vals, color=colors, width=0.5, edgecolor='white')
ax.bar_label(bars, labels=[f'{v*100:.1f}%' for v in vals],
             padding=3, fontsize=13, fontweight='bold', label_type='center', color='white')
ax.set_ylim(0, 1); ax.set_yticks([])
ax.spines[['top', 'right', 'left']].set_visible(False)
ax.set_title('Distribución de supervivencia', fontsize=13, fontweight='bold', color='#1E2761')
plt.tight_layout(); plt.show()
```

```python
#Diapositiva 2: Supervivencia por sexo
fig, axes = plt.subplots(1, 2, figsize=(9, 4))

surv_sex = titanic.groupby('sex')['survived'].mean().reindex(['male', 'female'])
axes[0].bar(['Hombres', 'Mujeres'], surv_sex.values, color=['#64748B', '#2A9D8F'],
            width=0.5, edgecolor='white')
for i, v in enumerate(surv_sex.values):
    axes[0].text(i, v/2, f'{v*100:.1f}%', ha='center', va='center',
                 fontweight='bold', fontsize=13, color='white')
axes[0].set_ylim(0, 1); axes[0].set_yticks([])
axes[0].spines[['top', 'right', 'left']].set_visible(False)
axes[0].set_title('Supervivencia por sexo', fontweight='bold', color='#1E2761')

surv_who = titanic.groupby('who')['survived'].mean().reindex(['man', 'child', 'woman'])
axes[1].bar(['Hombres', 'Niños', 'Mujeres'], surv_who.values,
            color=['#64748B', '#F4A261', '#2A9D8F'], width=0.5, edgecolor='white')
for i, v in enumerate(surv_who.values):
    axes[1].text(i, v/2, f'{v*100:.1f}%', ha='center', va='center',
                 fontweight='bold', fontsize=13, color='white')
axes[1].set_ylim(0, 1); axes[1].set_yticks([])
axes[1].spines[['top', 'right', 'left']].set_visible(False)
axes[1].set_title('Supervivencia por grupo', fontweight='bold', color='#1E2761')

plt.suptitle('"Las mujeres y los niños primero"', fontsize=14, fontweight='bold', color='#1E2761')
plt.tight_layout(); plt.show()
```

```python
#Diapositiva 3: Supervivencia por clase
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

#Global por clase
surv_cls = titanic.groupby('pclass')['survived'].mean()
colors_cls = ['#2A9D8F', '#F4A261', '#E63946']
bars = axes[0].bar(['1ª Clase', '2ª Clase', '3ª Clase'], surv_cls.values,
                   color=colors_cls, width=0.5, edgecolor='white')
for i, v in enumerate(surv_cls.values):
    axes[0].text(i, v/2, f'{v*100:.1f}%', ha='center', va='center',
                 fontweight='bold', fontsize=13, color='white')
axes[0].set_ylim(0, 1); axes[0].set_yticks([])
axes[0].spines[['top', 'right', 'left']].set_visible(False)
axes[0].set_title('Supervivencia por clase', fontweight='bold', color='#1E2761')

#Sexo x clase
pivot = titanic.pivot_table('survived', index='pclass', columns='sex', aggfunc='mean')
x = np.arange(3)
w = 0.35
b1 = axes[1].bar(x - w/2, pivot['female'], w, label='Mujeres', color='#2A9D8F', edgecolor='white')
b2 = axes[1].bar(x + w/2, pivot['male'],   w, label='Hombres', color='#64748B', edgecolor='white')
for bars_group in [b1, b2]:
    for bar in bars_group:
        h = bar.get_height()
        axes[1].text(bar.get_x() + bar.get_width()/2, h/2, f'{h*100:.0f}%',
                     ha='center', va='center', fontsize=9, fontweight='bold', color='white')
axes[1].set_xticks(x); axes[1].set_xticklabels(['1ª Clase', '2ª Clase', '3ª Clase'])
axes[1].set_ylim(0, 1.1); axes[1].set_yticks([])
axes[1].spines[['top', 'right', 'left']].set_visible(False)
axes[1].legend(fontsize=10); axes[1].set_title('Supervivencia por sexo y clase', fontweight='bold', color='#1E2761')

plt.suptitle('La clase del billete, un factor decisivo', fontsize=14, fontweight='bold', color='#1E2761')
plt.tight_layout(); plt.show()
```

```python
#Diapositiva 4: who x clase
pivot2 = titanic.pivot_table('survived', index='pclass', columns='who', aggfunc='mean')
fig, ax = plt.subplots(figsize=(9, 5))
x = np.arange(3)
w = 0.25
colors3 = ['#E63946', '#1E2761', '#2A9D8F']
labels3 = ['Niños', 'Hombres', 'Mujeres']
for i, (col, lbl, col_c) in enumerate(zip(['child', 'man', 'woman'], labels3, colors3)):
    bars = ax.bar(x + (i-1)*w, pivot2[col], w, label=lbl, color=col_c, edgecolor='white')
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h/2, f'{h*100:.0f}%',
                ha='center', va='center', fontsize=9, fontweight='bold', color='white')
ax.set_xticks(x); ax.set_xticklabels(['1ª Clase', '2ª Clase', '3ª Clase'], fontsize=11)
ax.set_ylim(0, 1.15); ax.set_yticks([])
ax.spines[['top', 'right', 'left']].set_visible(False)
ax.legend(fontsize=10)
ax.set_title('La clase explica las diferencias entre grupos\n(supervivencia por grupo y clase)',
             fontsize=13, fontweight='bold', color='#1E2761')
plt.tight_layout(); plt.show()
```

```python
#Diapositiva 5: Puerto de embarque
port_names = {'C': 'Cherburgo', 'Q': 'Queenstown', 'S': 'Southampton'}
surv_emb = titanic.groupby('embarked')['survived'].mean()
pct1_emb = titanic.groupby('embarked').apply(lambda x: (x['pclass'] == 1).mean())

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
colors5 = ['#CADCFC', '#64748B', '#1E2761']
labels5 = [port_names[k] for k in surv_emb.index]

for ax_i, (data, title) in enumerate([
    (surv_emb,  'Supervivencia por puerto'),
    (pct1_emb,  '% pasajeros de 1ª clase por puerto')
]):
    bars = axes[ax_i].bar(labels5, data.values, color=colors5, width=0.5, edgecolor='white')
    for bar, v in zip(bars, data.values):
        axes[ax_i].text(bar.get_x() + bar.get_width()/2, v/2, f'{v*100:.1f}%',
                        ha='center', va='center', fontweight='bold', fontsize=12,
                        color='black' if v < 0.3 else 'white')
    axes[ax_i].set_ylim(0, 0.85); axes[ax_i].set_yticks([])
    axes[ax_i].spines[['top', 'right', 'left']].set_visible(False)
    axes[ax_i].set_title(title, fontweight='bold', color='#1E2761', pad=8)
    axes[ax_i].tick_params(axis='x', rotation=15, labelsize=10)

plt.suptitle('El puerto embarca clase, no suerte', fontsize=14, fontweight='bold', color='#1E2761')
plt.tight_layout(); plt.show()
```
